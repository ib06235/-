```python
# 필요한 라이브러리, 모듈 import
import pandas as pd
```

```python
# 불러와 데이터프레임으로 변환
file_name = '인천도시공원정보표준데이터_3.csv'
df = pd.read_csv(file_name,encoding='cp949')
df.head(3)
```

```python
# 공원면적 1500이하 공원 삭제
idx_nm = df[df['공원면적'] < 1500].index
df = df.drop(idx_nm)
```

```python
# 시설설치가능면적 400 이하 공원 삭제
idx_nm = df[df['시설설치가능면적'] < 400].index
df = df.drop(idx_nm)
```

```python
# 데이터프레임을 csv로 저장
df.to_csv('인천_공원면적필터링.csv',index=False , sep=',', encoding='cp949')
```
