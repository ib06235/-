```python
# 필요한 라이브러리, 모듈 import
import pandas as pd
```

```python
# 불러와 데이터프레임으로 변환
file_name = '인천도시공원정보표준데이터_3.csv'
df = pd.read_csv(file_name,encoding='cp949')
df.head(3)
```

```python
# 시설 설치 가능 면적 to 공원 크기 점수로 변환 하는 함수
def 면적_to_점수(df):
    if df['시설설치가능면적'] >= 5000 :
        return 5
    elif df['시설설치가능면적'] >= 3500 :
        return 4
    elif df['시설설치가능면적'] >= 3000 :
        return 3
    elif df['시설설치가능면적'] >= 2000 :
        return 2
    elif df['시설설치가능면적'] >= 1500 :
        return 1
    else :
        return 0
```

```python
# 시설설치가능면적에 따라 계산된 새로운 컬럼(공원크기점수) 추가
df['공원크기점수'] = df.apply(면적_to_점수, axis=1)
```

```python
# 데이터프레임을 csv로 저장
df.to_csv('test4.csv',index=True , sep=',', encoding='cp949')
```
